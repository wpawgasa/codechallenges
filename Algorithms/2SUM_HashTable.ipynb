## 2-SUM

The goal of this problem is to implement a variant of the 2-SUM algorithm. The file contains 1 million integers, both positive and negative (there might be some repetitions!). This is your array of integers, with the $i_{th}$ row of the file specifying the $i_{th}$ entry of the array.

Your task is to compute the number of target values $t$ in the interval $[-10000,10000]$ (inclusive) such that there are distinct numbers $x,y$ in the input file that satisfy $x+y=t$









```python
input_file = 'algo1-programming_prob-2sum.txt'
low = -10000
high = 10000
```

```python
def load_nums(input_file):
    with open(input_file, 'r') as f:
        nums = [int(line.rstrip()) for line in f.readlines()]
    nums.sort()
    return nums

def two_sum(H, low, high):
    i = 0
    j = len(H)-1
    sumList = []
    count = 0
    while i < j:
        if H[i] + H[j] < low:
            i += 1
        elif H[i] + H[j] > high:
            j -= 1
        else:
            for pos in range(i+1, j+1):
                t = H[i] + H[pos]
                if t > high:
                    break
                elif H[i] != H[pos] and t >= low:
                    if t not in sumList:
                        sumList.append(t)
                        count += 1
                        print("Found x={},y={},t={}".format(H[i],H[pos],t))
            i += 1
    return count
```

```python
nums = load_nums(input_file)
H = {}
for i in range(len(nums)):
    H[i] = nums[i]

count_sum = two_sum(H, low, high)

```
